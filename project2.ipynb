```python
# Inner Join 1
# Joins city id, address, and postal code from pagila.address to country id from pagila.city based on the city ids matching and being greater than 350. 
# The rows are then ordered by city id in ascending order and only the first 10 are shown.
!psql -c "select address.city_id,address, postal_code, country_id from pagila.address inner join pagila.city on city.city_id=address.city_id where city.city_id>350 order by city_id limit 10"
```

```python
# Inner Join 2
# Joins film id, category id and actor id from pagila.film_category to actor id from pagila.film_actor based on the film ids matching and being greater than 50.
# The rows are then ordered by film id in ascending order and only the first 11 are shown.
!psql -c "select film_category.film_id, category_id, actor_id from pagila.film_category inner join pagila.film_actor on film_category.film_id=film_actor.film_id where film_category.film_id>50 order by film_category.film_id limit 11"
```

```python
# Inner Join 3
# Joins film id, title, description, and length from pagila.film to category id from pagila.film_category based on the film ids matching and having a length less than 90.
# The rows are then ordered by film id in descending order and only the first 11 are shown.
!psql -c "select film.film_id, title, description, category_id, length from pagila.film inner join pagila.film_category on film.film_id=film_category.film_id where length<90 order by film_category.film_id DESC limit 11"
```

```python
# Outer Join 1
# Joins rental id and rental date from pagila.rental to amount and payment date from pagila.payment based on matching rental ids being greater than 30.
# The rows are then ordered by amount in ascending order and only the first 11 are shown.
!psql -c "select rental.rental_id, rental_date, amount, payment_date from pagila.rental full outer join pagila.payment on rental.rental_id=payment.rental_id where rental.rental_id>30 order by amount limit 11"
```

```python
# Outer Join 2
# Joins staff id and email from pagila.staff to last update from pagila.store based on matching store ids being greater than 0.
# The rows are then ordered by email in descending order.
!psql -c "select staff.staff_id, email, store.last_update from pagila.staff full outer join pagila.store on staff.store_id=store.store_id where staff.staff_id>0 order by email DESC"
```

```python
# Outer Join 3
# Joins original language id, language id and replacement cost from pagila.film to language name from pagila.language based on original language id and language id from pagila.language matching.
# From there, rows where replacement cost is greater than 5 are kept.
# The rows are then ordered by repalcement cost in descending order and only 11 are chosen.
!psql -c "select original_language_id, film.language_id, replacement_cost, language.name from pagila.film full outer join pagila.language on language.language_id=original_language_id where replacement_cost>5 order by replacement_cost limit 11" 
```

```python
# Self Join
#joins two titles and the film length from pagila.film to display a table that shows all films that are the same length of one specific film ordered by the film length.
!psql -c "select film1.title title1, film2.title title2, film1.length film_length from pagila.film film1 inner join pagila.film film2 on film1.film_id<>film2.film_id and film1.length = film2.length order by film_length"
```
